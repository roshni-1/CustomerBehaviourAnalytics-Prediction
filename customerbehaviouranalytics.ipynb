```python
# Applying log transformation to highly skewed features
df['Age_Income_Ratio'] = np.log1p(df['Age_Income_Ratio'])  # Log transformation for Age_Income_Ratio
df['Online_Purchase_Tendency'] = np.sqrt(df['Online_Purchase_Tendency'])  # Square root transformation for online purchase tendency

```

```python
# Defining a function to cap outliers
def cap_outliers(data, cols, threshold=3):
    for col in cols:
        mean = data[col].mean()
        std = data[col].std()
        upper_limit = mean + threshold * std
        lower_limit = mean - threshold * std
        data[col] = np.clip(data[col], lower_limit, upper_limit)
    return data

# Applying outlier capping to numeric features
df = cap_outliers(df, numeric_columns)
```

```python
from sklearn.preprocessing import StandardScaler  # For scaling the data using standard scaling
from sklearn.model_selection import train_test_split # For splitting the dataset
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor # Using Random Forest classifier and regressor
from sklearn.metrics import classification_report, roc_auc_score, mean_squared_error, r2_score # Evaluation metrics for evaluating the model

# Selecting Best Features for Classification and Regression
# Classification: Retain top features based on importance
selected_features_classification = [
    'Age', 'Age_Income_Ratio', 'Purchases_in_Last_Month',
    'Income', 'Recency_Frequency_Score', 'Is_Young_Professional'
]

# Regression: Retaining top features based on importance
selected_features_regression = [
    'Income', 'Customer_Loyalty_Score', 'Purchases_in_Last_Month',
    'Days_Since_Last_Purchase', 'Online_Purchase_Tendency'
]

# Preparing Data for Modeling
# Classification Dataset
X_classification = df_encoded[selected_features_classification]
y_classification = df['Will_Purchase']

# Regression Dataset
X_regression = df_encoded[selected_features_regression]
y_regression = df['Annual_Spending']

# Scaling Features
scaler = StandardScaler()
X_classification_scaled = scaler.fit_transform(X_classification)
X_regression_scaled = scaler.fit_transform(X_regression)

# Splitting Dataset
# Classification
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_classification_scaled, y_classification, test_size=0.2, random_state=42, stratify=y_classification
)

# Regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regression_scaled, y_regression, test_size=0.2, random_state=42
)

# Step 5: Build and Evaluate Models

# Classification Model: Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train_cls, y_train_cls)
y_pred_cls = rf_clf.predict(X_test_cls)
y_proba_cls = rf_clf.predict_proba(X_test_cls)[:, 1]

# Classification Evaluation
classification_report_result = classification_report(y_test_cls, y_pred_cls)
roc_auc = roc_auc_score(y_test_cls, y_proba_cls)

# Regression Model: Random Forest Regressor
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train_reg, y_train_reg)
y_pred_reg = rf_reg.predict(X_test_reg)

# Regression Evaluation
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = mse ** 0.5
r2 = r2_score(y_test_reg, y_pred_reg)

# Displaying Results
{
    "Classification Report": classification_report_result,
    "ROC AUC Score": roc_auc,
    "Regression MSE": mse,
    "Regression RMSE": rmse,
    "Regression R2": r2
}
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initializing models for regression
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0, random_state=42),
    "Gradient Boosting Regressor": GradientBoostingRegressor(random_state=42, n_estimators=100, learning_rate=0.1, max_depth=3)
}

# Training and evaluating each model
results = {}
for name, model in models.items():
    # Training  model
    model.fit(X_train_reg, y_train_reg)

    # Making predictions
    y_pred = model.predict(X_test_reg)

    # Evaluating the model
    mse = mean_squared_error(y_test_reg, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_test_reg, y_pred)

    # Storing the results
    results[name] = {
        "MSE": mse,
        "RMSE": rmse,
        "R2": r2
    }
```
